```python
import numpy as np
import pandas as pd
from  sklearn.tree import DecisionTreeClassifier
```

```python
data=pd.read_csv('tit_train.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
drop_columns=['name','ticket','cabin','embarked','home.dest','boat','body']
```

```python
data_clean=data.drop(columns=drop_columns,axis=1)
```

```python
data_clean.info()
data_clean.head(n=10)
```

```python
#Since the sex is a string first of all encode it to label 0 and 1
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
data_clean['sex']=le.fit_transform(data_clean['sex'])
data_clean.head()
```

```python
data_clean=data_clean.fillna(data_clean['age'].mean())
data_clean.head()
```

```python
input_cols=['pclass','sex','age','sibsp','parch','fare']
output_cols=['survived']
X=data_clean[input_cols]
Y=data_clean[output_cols]
X.head()
```

```python
Y.head()
```

```python
Xtest=pd.read_csv('tit_test.csv')
drop_columns=['name','ticket','cabin','embarked','boat','body','home.dest']
data_test_clean=Xtest.drop(columns=drop_columns,axis=1)
data_test_clean.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data_test_clean['sex']=le.fit_transform(data_test_clean['sex'])
data_test_clean=data_test_clean.fillna(data_test_clean['age'].mean())
data_test_clean.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfk = RandomForestClassifier(criterion='entropy', max_depth=6)
rfk.fit(X,Y.values.ravel())
```

```python
rfk.score(X,Y)
```

```python
y=rfk.predict(data_test_clean)
```

```python
print(y)
```

```python
x=y.astype(int)
```

```python
df = pd.DataFrame(data=x , columns=["survived"])
df.to_csv("Titanic_Challenge.csv",index=True,index_label ="Id")
```
